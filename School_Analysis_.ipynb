5. If the marks obtained for each subject can be categorised under 5 sections like below:
    
     <=20 - Very Poor
    
    >20 - <=40 - Poor
    
    >40 - <=60 - Average
    
    >60 - <=80 - Good
    
    >80 - <=100 - Very Good
    

Then calculate for each school how many students were in each category based on the avg. marks obtained each year

```python
def categorize_marks(marks):
    if marks <= 20:
        return 'Very Poor'
    elif marks <= 40:
        return 'Poor'
    elif marks <= 60:
        return 'Average'
    elif marks <= 80:
        return 'Good'
    else:
        return 'Very Good'
```

```python
data['Average Marks'] = data.iloc[:, 2:13].mean(axis=1)
data['Category'] = data['Average Marks'].apply(categorize_marks)
```

```python
category_counts = data.groupby('Year')['Category'].value_counts()
print("Category Counts Per Year:")
print(category_counts)
```

```python
for year in [2019, 2020, 2021]:
    category_data = category_counts.loc[year]
    plt.pie(category_data, labels=category_data.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Student Distribution in Categories for {year}')
    plt.show()
```

Explanation:-The pie chart will display the distribution of students across different categories (Very Poor, Poor, Average, Good, Very Good) for each year (2019, 2020, and 2021). The chart will show the percentage of students in each category based on their average marks for each year.

7.

Which is the fastest-growing School in Bangalore (Overall and Streamwise)?

```python
# Define streams and associated subjects
streams = {
    'Arts': ['Hindi', 'English', 'History', 'Geography', 'Civics'],
    'Science': ['Mathematics', 'Physics', 'Chemistry', 'Biology', 'Computer Science'],
    'Commerce': ['Hindi', 'English', 'Mathematics', 'Computer Science', 'Physical Education']
}
```

```python
# Add a Total Marks column by summing all subject columns
subject_columns = ['Hindi', 'English', 'Mathematics', 'Physics', 'Chemistry',
                   'Biology', 'History', 'Geography', 'Civics', 'Computer Science',
                   'Physical Education']
```

```python

# Sum across the subject columns to calculate Total Marks
data['Total Marks'] = data[subject_columns].sum(axis=1)
```

```python
data.head()
```

```python
data.isnull().sum()
data['Year'].unique()  # Ensure all years (2019, 2020, 2021) are present
```

```python
data.fillna(0, inplace=True)
```

```python
data.groupby(['School', 'Year'])['Total Marks'].count()
```

```python
# Pivot data to calculate growth
growth = data.groupby(['School', 'Year'])['Total Marks'].mean().unstack()
growth['Growth'] = growth[2021] - growth[2019]

# Identify the fastest-growing school overall
fastest_growing_school = growth['Growth'].idxmax()
print("Fastest-Growing School Overall:")
print(fastest_growing_school)

# Streamwise growth
stream_growth = {}
for stream, subjects in streams.items():
    stream_data = data.groupby(['School', 'Year'])[subjects].mean()
    stream_data = stream_data.sum(axis=1).unstack()
    stream_data['Growth'] = stream_data[2021] - stream_data[2019]
    stream_growth[stream] = stream_data['Growth'].idxmax()

print("\nFastest-Growing Schools Streamwise:")
print(stream_growth)
```
